### move-zeroes

Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.

```python
def moveZeroes(nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        len_nums = len(nums)
        k=0
        for i in range(len_nums-1,-1,-1):
                if nums[i] == 0: 
                       nums.remove(nums[i])
                       k+=1
        nums.extend(k*[0])
        return nums   
```

```python
nums = [0,1,0,3,12]
moveZeroes(nums)
```

```python
def moveZeroes(nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        start=0
        stop=len(nums)
        count=0
        while start<stop:
                if nums[start]==0:
                        nums.pop(start)
                        count+=1
                stop-=1
                start+=1
        nums.extend([0]*count)
        return nums     
```

```python
nums = [0,1,0,3,12]
moveZeroes(nums)
```
